## Conditional Statements (if, elif, else)
Conditional statements allow you to execute different blocks of code based on specific conditions. The most commonly used conditional statements are `if`, `elif`, and `else`.


```python

# Example 1: Age Group Classification
age = 20

if age >= 18:
    print("You are an adult.")
elif age >= 13:
    print("You are a teenager.")
else:
    print("You are a child.")
```

```python
# Example 2
num = int(input("Enter a number: "))
if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
```

## Loops (for, while)
Loops allow you to execute a block of code multiple times. Python provides two types of loops: `for` loops and `while` loops.


```python

# Example of `for` loop
for i in range(5):
    print(f"Iteration {i}")
```

```python

# Example of Sum of Numbers using `for` loop
numbers = [5, 10, 15, 20]
total = 0
for num in numbers:
    total += num
print(f"The sum of the numbers is: {total}")
```

## Functions (with return statements affecting control flow)

Functions are reusable blocks of code that perform specific tasks.

The `return` statement is used to return a result from the function, and it affects control flow by ending the function’s execution and passing the result back.

```python
# Example

def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

```python

# Example of Function with return
def square(num):
    return num ** 2

result = square(4)
print(f"The square of 4 is {result}")
```

### More Built-in Functions

```python

# Example: Using `len()` function to find the length of a string
text = "Hello, world!"
length = len(text)
print(f"The length of the text is: {length}")
```

```python

# Example: Using `max()` and `min()` to find the largest and smallest in a list
numbers = [5, 3, 9, 1, 7]
max_value = max(numbers)
min_value = min(numbers)
print(f"The maximum value is: {max_value}")
print(f"The minimum value is: {min_value}")
```

```python

# Example: Using `sum()` function to find the sum of a list
numbers = [10, 20, 30, 40]
total_sum = sum(numbers)
print(f"The sum of the numbers is: {total_sum}")
```

```python

# Example: Using `abs()` function to find the absolute value of a number
num = -10
absolute_value = abs(num)
print(f"The absolute value of {num} is {absolute_value}")
```

## Data Structures: Lists and Dictionaries
Python provides built-in data structures such as lists and dictionaries that are essential for storing and manipulating data.


```python

# Example: Using Lists
fruits = ["apple", "banana", "cherry"]
print(f"The first fruit is {fruits[0]}")
fruits.append("orange")
print(f"Updated list: {fruits}")
```

```python

# Example: Using Dictionaries
student = {"name": "John", "age": 21, "course": "Computer Science"}
print(f"Student name: {student['name']}")
student["age"] = 22  # Update age
print(f"Updated student: {student}")
```

```python

# Example: List Operations
numbers = [1, 2, 3, 4, 5]
print(f"First three numbers: {numbers[:3]}")
for num in numbers:
    print(f"Number: {num}")
```

```python

# Example: Nested List
students = [["John", 21], ["Alice", 22], ["Bob", 20]]
print(f"First student: {students[0]}")
```

```python

# Example: Nested Dictionary
students_info = {
    "John": {"age": 21, "course": "Computer Science"},
    "Alice": {"age": 22, "course": "Engineering"}
}
print(f"John's course: {students_info['John']['course']}")
```
